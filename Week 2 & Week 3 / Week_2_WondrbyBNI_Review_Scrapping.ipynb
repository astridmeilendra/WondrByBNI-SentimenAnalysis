```python
from google_play_scraper import app
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as dates
import seaborn as sns
```

Data Scrapping

```python
from google_play_scraper import app, Sort, reviews_all

wondr_reviews = reviews_all(
    'id.bni.wondr',  # Google Play id untuk wondr by BNI
    sleep_milliseconds=0,
    lang='id',  # bahasa review
    sort=Sort.NEWEST,
)
```

```python
# Save the Wondr app reviews into dataframe
df_wondrRev = pd.DataFrame(np.array(wondr_reviews), columns=['content'])
df_wondrRev = df_wondrRev.join(pd.DataFrame(df_wondrRev.pop('content').tolist()))
df_wondrRev.to_csv(r'df_wondrRev.csv', index=False)
df_wondrRev.head()
```
