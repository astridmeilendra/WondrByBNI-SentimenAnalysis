```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

```

```python
# 1. Dataset dari slide Stanford
corpus = [
    "Banjir yang melanda sejumlah daerah menyebabkan ribuan rumah warga terendam, sehingga evakuasi warga harus segera dilakukan ke lokasi yang lebih aman.",
    "Curah hujan tinggi dalam beberapa hari terakhir memicu banjir yang meluas, sehingga proses evakuasi warga terkendala akibat akses jalan yang ikut terendam.",
    "Pemerintah daerah bersama tim tanggap darurat terus menangani banjir dengan mempercepat evakuasi korban serta menyalurkan bantuan bagi para pengungsi."
]

# 2. Representasi Vektor (Hanya fokus pada kata 'hate' dan 'darkness')
vectorizer = CountVectorizer(vocabulary=['banjir', 'evakuasi'])
X = vectorizer.fit_transform(corpus)

```

```python
# 3. Hitung Cosine Similarity
# Membandingkan semua dokumen dengan Dokumen 0 (1 John)
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan Hasil
df = pd.DataFrame(X.toarray(), columns=['banjir', 'evakuasi'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# 1. Dataset dari slide Stanford
corpus = [
    "Whoever has hate for his brother is in the darkness and walks in the darkness.", # 1 John
    "Hello darkness, my old friend.",                                                  # S&G
    "Returning hate for hate multiplies hate, adding deeper darkness to a night already devoid of stars. Darkness cannot drive out darkness; only light can do that." # MLK
]

# 2. Representasi Vektor (Hanya fokus pada kata 'hate' dan 'darkness')
vectorizer = CountVectorizer(vocabulary=['hate', 'darkness'])
X = vectorizer.fit_transform(corpus)

# 3. Hitung Cosine Similarity
# Membandingkan semua dokumen dengan Dokumen 0 (1 John)
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan Hasil
df = pd.DataFrame(X.toarray(), columns=['hate', 'darkness'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

**CountVectorizer:** Mengubah teks menjadi angka (matriks frekuensi kata). Di sini kita membatasi fitur hanya pada kata "hate" dan "darkness" agar sesuai dengan grafik di slide Stanford.

**Cosine Similarity vs Distance:** Fungsi ini menghasilkan nilai 0 sampai 1. Semakin mendekati 1, semakin mirip kedua dokumen tersebut.

**Analisis Hasil:** Bahwa Dokumen 2 (MLK) memiliki nilai similarity yang tinggi terhadap Dokumen 0, membuktikan bahwa meskipun jumlah katanya jauh lebih banyak, "arah" topiknya tetap serupa.

**Euclidean Distance:** Jika kita menggunakan jarak lurus, Dokumen 2 akan dianggap sangat berbeda karena jumlah kemunculan katanya (3 dan 4) jauh dari Dokumen 0 (1 dan 2).

**Aplikasi di Dunia Nyata:** Teknik ini adalah dasar dari mesin pencari (search engines) dan sistem rekomendasi teks yang sering kita gunakan sehari-hari.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# 1. Dataset dari slide Stanford
corpus = [
    "Banjir yang melanda sejumlah daerah menyebabkan ribuan rumah warga terendam, sehingga evakuasi warga harus segera dilakukan ke lokasi yang lebih aman.",
    "Curah hujan tinggi dalam beberapa hari terakhir memicu banjir yang meluas, sehingga proses evakuasi warga terkendala akibat akses jalan yang ikut terendam.",
    "Pemerintah daerah bersama tim tanggap darurat terus menangani banjir dengan mempercepat evakuasi korban serta menyalurkan bantuan bagi para pengungsi."
]

# 2. Representasi Vektor (Hanya fokus pada kata 'hate' dan 'darkness')
vectorizer = CountVectorizer(vocabulary=['banjir', 'terendam'])
X = vectorizer.fit_transform(corpus)

# 3. Cosine Similarity
cos_sim = cosine_similarity(X[0:1], X)

# 4. Tampilkan hasil
df = pd.DataFrame(X.toarray(), columns=['banjir', 'terendam'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()

print(df)
```

```python
import numpy as np
def length(v):
    return np.sqrt((v ** 2).sum())
def cos_dist(v, w):
    return 1 - (v * w).sum() / (length(v) * length(w))
cos_dist(tf.loc[0], tf.loc[1]), cos_dist(tf.loc[0], tf.loc[2])
```
